Task 2 : Classify Iris FLower

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

Loading the Dataset

```python
columns = ['Sepal lengeth', 'Sepal width', 'Petal length', 'Petal width', 'Class_labels']
#load the data
df = pd.read_csv('iris.data', names=columns)
df.head(50)
```

Data Visualization

```python
df.describe()
```

```python
#visualize the whole dataset
sns.pairplot(df, hue='Class_labels')
```

Separating input columns and the output columns

```python
#separate features and targets
data = df.values

X = data[:,0:4] #last colummn is removed
Y = data[:,4]   #only shows the last column
```

Splitting the data into training and testing data / dividing them

```python
#Split the dataset into train and test dataset
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X, Y, test_size = 0.2)
```

Model 1: Support Vector Machine Algoritm

```python
#support vector machine algoritm
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(X_train, y_train)
```

```python
prediction1 = model_svc.predict(X_test)
#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction1)*100)
for i in range(len(prediction1)):
  print(y_test[i],prediction1[i])  #comparing values by accuracy
```

Model 2: Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model_LR = LogisticRegression()
model_LR.fit(X_train, y_train)
```

```python
prediction2 = model_LR.predict(X_test)
#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction2)*100)
```

Model 3: Decision Tree Classifier

```python
#DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
model_DTC = DecisionTreeClassifier()
model_DTC.fit(X_train,y_train)
```

```python
prediction3 = model_DTC.predict(X_test)
#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction3)*100)
```

```python
#a detailed classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, prediction1))
```
